# Imputing

### Gender

```python
print(df[['Gender']].mode())
```

```python
df['Gender']=df['Gender'].fillna('Male')
print(df['Gender'].unique())
print(df['Gender'])
print("Missing Values", df['Gender'].isnull().sum())
```

### Dependents

```python
print(df['Dependents'].isnull().sum())
print(df['Dependents'].mode())
```

```python
df[['Dependents']].mode()
```

```python
df['Dependents']=df['Dependents'].fillna('0')
print(df[['Dependents']])
print("Missing Values",df['Dependents'].isnull().sum())
```

### Education

```python
print("Missing Values",df['Education'].isnull().sum())
```

### Self_Employed

```python
print("Missing Values",df['Self_Employed'].isnull().sum())
print(df[['Self_Employed']].mode())
```

```python
df['Self_Employed']=df['Self_Employed'].fillna("No")
print(df[['Self_Employed']])
print("Missing Values",df['Self_Employed'].isnull().sum())
```

### Applicant Income

```python
print("Missing Values",df['ApplicantIncome'].isnull().sum())
```

### Co-Applicant Income

```python
print("Missing Values",df['CoapplicantIncome'].isnull().sum())
```

### LoanAmount

```python
print("Missing Values",df['LoanAmount'].isnull().sum())
```

```python
df[['LoanAmount']].mean()
```

```python
df['LoanAmount']=df['LoanAmount'].fillna(146.412162)
print(df[['LoanAmount']])
print("Missing Values",df['LoanAmount'].isnull().sum())
```

### Loan Amount Term

```python
print("Missing Values",df['Loan_Amount_Term'].isnull().sum())
```

```python
print(df['Loan_Amount_Term'].mean())
df['Loan_Amount_Term']=df['Loan_Amount_Term'].fillna(342.0)
print(df[['Loan_Amount_Term']])
print("Missing Values",df['Loan_Amount_Term'].isnull().sum())
```

### Credit History

```python
print("Missing Values",df['Credit_History'].isnull().sum())
```

```python
print(df['Credit_History'].mean())
df['Credit_History']=df['Credit_History'].fillna(342.0)
print(df[['Credit_History']])
print("Missing Values",df['Credit_History'].isnull().sum())
```
